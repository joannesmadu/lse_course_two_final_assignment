```python
# Create a new DataFrame containing only the text.

tweets_text = tweets[['tweet_full_text']]

# View the DataFrame.

tweets_text
```

```python
# Loop through the messages, and create a list of values containing the # symbol.

tags = []

for tokenized in [word.split(' ') for word in tweets_text['tweet_full_text'].values]:
    for item in tokenized:
        if '#' in item:
            # Change to lowercase.
            tags.append(item.lower())
```

```python
tags
```

```python
# Display the first 30 records.

tags = pd.Series(tags, name='tags')
tags.head(30)
```

```python
# Convert the series to a DataFrame in preparation for visualisation.

tags_data = tags.to_frame().reset_index()
tags_data

# Rename the columns.
tags_data = tags_data.rename(columns={'index':'count', 'tags':'word'})
tags_data
```

```python
# Fix the count datatype.

grouped = tags_data.groupby(['word'])

# View the result.

tags_data_grouped = grouped.sum().reset_index()
tags_data_grouped
```

```python
# Display records where the count is larger than 10.

count_over_ten = tags_data_grouped[tags_data_grouped['count'] > 10]
count_over_ten
```

```python
# Create a Seaborn barplot displaying records with a count larger than 10.

sns.barplot(
    x='word',
    y='count',
    data=count_over_ten,
    errorbar=None)
plt.title('Tweets: Record Above 10')
plt.xticks(rotation=45)
plt.xlabel('Hashtags')
plt.ylabel('Frequency')
plt.legend(title='Hashtags')
plt.show()
```

```python
# Create the plot.
top_tweets = tags_data_grouped.sort_values('count', ascending=False).head(30)
top_tweets

# View the barplot.

sns.barplot(
    x='word',
    y='count',
    data=top_tweets)
plt.title('Tweets: Top 30 By Frequency')
plt.xticks(rotation=75)
plt.xlabel('Hashtags')
plt.ylabel('Frequency')
plt.show()
```
